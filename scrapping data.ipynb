```python
name = box_1['teams']['away']['team']['name']
home = 'yes'
goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['goals']
shots = pim = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['shots']
blocked_shots = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['blocked']
pim = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['pim']
pp_percent = pim = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['powerPlayPercentage']   
pp_goals = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['powerPlayGoals']
pp_chances = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['powerPlayOpportunities']
faceoff_percent = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['faceOffWinPercentage']
takeaways = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['takeaways']
giveaways = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['giveaways']
hits = goals = box_1['teams']['away']['teamStats']['teamSkaterStats']['hits']
```

```python
game_1_box_df = pd.DataFrame()
i = box_1_pull.json()
for location in ['home', 'away']:
    if location == 'home':
        score_reverser = 'away'
    if location == 'away':
        score_reverser = 'home'
    name = i['teams'][location]['team']['name']
    home_status = location
    goals = i['teams'][location]['teamStats']['teamSkaterStats']['goals']
    shots = i['teams'][location]['teamStats']['teamSkaterStats']['shots']
    blocked_shots = i['teams'][location]['teamStats']['teamSkaterStats']['blocked']
    pims = goals = i['teams'][location]['teamStats']['teamSkaterStats']['pim']
    pp_percent = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayPercentage']
    pp_goals = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayGoals']
    pp_chances = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayOpportunities']
    faceoff_percent = i['teams'][location]['teamStats']['teamSkaterStats']['faceOffWinPercentage']
    takeaways = i['teams'][location]['teamStats']['teamSkaterStats']['takeaways']
    giveaways = i['teams'][location]['teamStats']['teamSkaterStats']['giveaways']
    hits = i['teams'][location]['teamStats']['teamSkaterStats']['hits']
    goals_diff = (i['teams'][location]['teamStats']['teamSkaterStats']['goals']) - \
        i['teams'][score_reverser]['teamStats']['teamSkaterStats']['goals']
    
    temp_dic = {
    'team': name, 'home_status': home_status, 'goals': goals, 'shots': shots, 'blocked_shots': blocked_shots,
    'penalty_minutes': pims, 'power_play_percentage': pp_percent, 'power_play_goals': pp_goals,
    'power_play_chances': pp_chances, 'faceoff_percent':faceoff_percent,'takeaways':takeaways,
    'giveaways' : giveaways,'hits':hits, 'goal_diff':goals_diff
    }
    temp_df = pd.DataFrame(temp_dic, index=[0])
    temp_df['won'] = temp_df['goal_diff'].map(lambda x: 'yes' if x>0 else 'no')
    game_1_box_df = pd.concat([game_1_box_df, temp_df], axis=0, ignore_index=True)
```
