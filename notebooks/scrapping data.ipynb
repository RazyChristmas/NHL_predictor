```python
# loop that pulls down all the games from the season 2020-2021 and returns a list of type request results
big_pull = []
for i in range(1,1272):
    if i <10:
        game_id = f'000{i}'
    if i >= 10 and i<100:
        game_id = f'00{i}'
    if i >= 100 and i <1000:
        game_id = f'0{i}'
    if i >= 1000:
        game_id=f'{i}'
    game_pull = requests.get(f'https://statsapi.web.nhl.com/api/v1/game/202002{game_id}/feed/live')
    if game_pull.status_code == 200:
        big_pull.append(game_pull)
    
    time.sleep(1)
```

Pulling out every event for every game in the list. Using all of the data at the same time was very memory intensive, so to save time I ran it in chunks of 100 games. Each set of 100 pulls is converted into a dataframe, the several dataframes are added together at the end, resulting in one big table with every game event from the season 2020-2021

```python
big_data_0_100 = pd.DataFrame()
a = 0
for i in range(101):
    i = big_pull[i].json()
    gameplay = i['liveData']['plays']['allPlays']
    for n in range(3,len(gameplay),1):
        event = gameplay[n]['result']['event']
        try:
            team = gameplay[n]['team']['name']
        except:
            team = 'NaN'
        try:
            x_cords = gameplay[n]['coordinates']['x']
        except:
            x_cords = "NaN"
        try:
            y_cords = gameplay[n]['coordinates']['y']
        except:
            y_cords = 'NaN'
        period = gameplay[n]['about']['period']
        periodTime = gameplay[n]['about']['periodTime']
        game_score = gameplay[n]['about']['goals']

        home_team = i['gameData']['teams']['home']['name']
        game_id = i['gameData']['game']['pk']
        temp_dic = {'event': event, 'team': team,'x_coordinate':x_cords, 'y_coordinate': y_cords,
                   'period':period, 'period_time': periodTime, 'game_id':game_id,'score': game_score,
               'home_team': home_team}
        temp_df = pd.DataFrame(temp_dic, index=[n])
        big_data_0_100 = pd.concat([big_data_0_100, temp_df], axis=0, ignore_index=True)
    print(f'game{a} done')
    a += 1
```

game box scores, these will be the aggregate stats at the end of every game for the season 2020-20201. 

```python
#same pull loop as before, using different API hook to get boxscores. 
box_pull = []
for i in range(1,880):
    if i <10:
        game_id = f'000{i}'
    if i >= 10 and i<100:
        game_id = f'00{i}'
    if i >= 100 and i <1000:
        game_id = f'0{i}'
    if i >= 1000:
        game_id=f'{i}'
    game_pull = requests.get(f'https://statsapi.web.nhl.com/api/v1/game/202002{game_id}/boxscore')
    if game_pull.status_code == 200:
        box_pull.append(game_pull)
    if i in [100,200,300,400,500,600,700,800]:
        print(f'loop {i} done')
    
    time.sleep(1)
```

```python
big_box = pd.DataFrame()
a = 0
for i in box_pull:
    i = i.json()
    for location in ['away', 'home']:
        if location == 'home':
            score_reverser = 'away'
        if location == 'away':
            score_reverser = 'home'
        name = i['teams'][location]['team']['name']
        home_status = location
        goals = i['teams'][location]['teamStats']['teamSkaterStats']['goals']
        shots = i['teams'][location]['teamStats']['teamSkaterStats']['shots']
        blocked_shots = i['teams'][location]['teamStats']['teamSkaterStats']['blocked']
        pim = i['teams'][location]['teamStats']['teamSkaterStats']['pim']
        pp_percent = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayPercentage']
        pp_goals = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayGoals']
        pp_chances = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayOpportunities']
        faceoff_percent = i['teams'][location]['teamStats']['teamSkaterStats']['faceOffWinPercentage']
        takeaways = i['teams'][location]['teamStats']['teamSkaterStats']['takeaways']
        giveaways = i['teams'][location]['teamStats']['teamSkaterStats']['giveaways']
        hits = i['teams'][location]['teamStats']['teamSkaterStats']['hits']
        goals_diff = (i['teams'][location]['teamStats']['teamSkaterStats']['goals']) - \
            i['teams'][score_reverser]['teamStats']['teamSkaterStats']['goals']
        temp_dic = {
            'team': name, 'home_status': home_status, 'goals': goals, 'shots': shots, 'blocked_shots': blocked_shots,
            'penalty_minutes': pim, 'power_play_percentage': pp_percent, 'power_play_goals': pp_goals,
            'power_play_chances': pp_chances, 'faceoff_percent':faceoff_percent,'takeaways':takeaways,
            'giveaways' : giveaways,'hits':hits, 'goal_diff':goals_diff
            }
        temp_df = pd.DataFrame(temp_dic, index=[0])
        temp_df['won'] = temp_df['goal_diff'].map(lambda x: 'yes' if x>0 else 'no')
        big_box = pd.concat([big_box, temp_df], axis=0, ignore_index=True)
    #print(f'loop {a} done')
    a += 1
    
```

```python
#function to pull a list of years box scores, just in case I need extra box score data. 
def pull_box_scores(years):
    """Function that takes in a list of years and returns the full season box score stats. Use the
    first year of the season to denote it (IE: 2020 for the 2020-2021 season). If playoffs is True then 
    only playoff games from that year will be returned. requries time, request, and pandas libraries """
    pull_list = []
    game_state = '02'
    box_scores = pd.DataFrame()
    for year in years:
        print (str(year) +' starting')
        for i in range(1,1271):
            if i <10:
                game_id = f'000{i}'
            if i >= 10 and i<100:
                game_id = f'00{i}'
            if i >= 100 and i <1000:
                game_id = f'0{i}'
            if i >= 1000:
                game_id=f'{i}'
            game_pull = requests.get(f'https://statsapi.web.nhl.com/api/v1/game/{str(year)}{game_state}{game_id}/boxscore')
            if game_pull.status_code == 200:
                pull_list.append(game_pull)
            if i in list(range(0,1300,200)):
                print(f'loop {i} done')
            time.sleep(1)
        print (str(year) +' done')
    for i in pull_list:
        i = i.json()
        for location in ['away', 'home']:
            if location == 'home':
                score_reverser = 'away'
            if location == 'away':
                score_reverser = 'home'
            name = i['teams'][location]['team']['name']
            home_status = location
            goals = i['teams'][location]['teamStats']['teamSkaterStats']['goals']
            shots = i['teams'][location]['teamStats']['teamSkaterStats']['shots']
            blocked_shots = i['teams'][location]['teamStats']['teamSkaterStats']['blocked']
            pim = i['teams'][location]['teamStats']['teamSkaterStats']['pim']
            pp_percent = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayPercentage']
            pp_goals = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayGoals']
            pp_chances = i['teams'][location]['teamStats']['teamSkaterStats']['powerPlayOpportunities']
            faceoff_percent = i['teams'][location]['teamStats']['teamSkaterStats']['faceOffWinPercentage']
            takeaways = i['teams'][location]['teamStats']['teamSkaterStats']['takeaways']
            giveaways = i['teams'][location]['teamStats']['teamSkaterStats']['giveaways']
            hits = i['teams'][location]['teamStats']['teamSkaterStats']['hits']
            goals_diff = (i['teams'][location]['teamStats']['teamSkaterStats']['goals']) - \
                i['teams'][score_reverser]['teamStats']['teamSkaterStats']['goals']
            temp_dic = {
                'team': name, 'home_status': home_status, 'goals': goals, 'shots': shots, 'blocked_shots': blocked_shots,
                'penalty_minutes': pim, 'power_play_percentage': pp_percent, 'power_play_goals': pp_goals,
                'power_play_chances': pp_chances, 'faceoff_percent':faceoff_percent,'takeaways':takeaways,
                'giveaways' : giveaways,'hits':hits, 'goal_diff':goals_diff
                }
            temp_df = pd.DataFrame(temp_dic, index=[0])
            temp_df['won'] = temp_df['goal_diff'].map(lambda x: 'yes' if x>0 else 'no')
            box_scores = pd.concat([box_scores, temp_df], axis=0, ignore_index=True)
    return box_scores
```
